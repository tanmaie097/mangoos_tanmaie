# pandas vs spark JOIN

pandas

```python
import pandas as pd
data = pd.read_csv("/Users/tanmaie/Desktop/magnoos/exercises/01_pandas/bigmart_data.csv")
data.columns
```

```python
items = data[['Item_Identifier','Item_Weight','Item_Fat_Content','Item_Visibility','Item_Type','Item_MRP',]]
outlets = data[['Item_Identifier','Outlet_Identifier','Outlet_Establishment_Year','Outlet_Size','Outlet_Location_Type','Outlet_Type','Item_Outlet_Sales']]
```

```python
inner = pd.merge(items, outlets, on = 'Item_Identifier',how = 'inner')
inner.size
```

```python
left = pd.merge(items, outlets, on = 'Item_Identifier',how = 'left')
left.size
```

```python
outer = pd.merge(items, outlets, on = 'Item_Identifier',how = 'outer')
outer.size
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("test_02").getOrCreate()
```

```python
print(type(data))
```

```python
csv_path = "/Users/tanmaie/Desktop/magnoos/exercises/01_pandas/bigmart_data.csv"
data_spark = spark.read.option("header", "true").option("inferSchema", "true").csv(csv_path)
data_spark.show(5)
```

```python
items = data_spark.select('Item_Identifier','Item_Weight','Item_Fat_Content','Item_Visibility','Item_Type','Item_MRP')
outlets = data_spark.select('Item_Identifier','Outlet_Identifier','Outlet_Establishment_Year','Outlet_Size','Outlet_Location_Type','Outlet_Type','Item_Outlet_Sales')

# JOIN on Item_Identifier (same as pandas merge)
joined_df = items.join(outlets, on="Item_Identifier", how="inner")

joined_df.show(10)
```
